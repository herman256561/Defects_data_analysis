```python
import pandas as pd
import matplotlib.pyplot as plt
import squarify

# Load your actual dataset
file_path = 'defects_data.csv'  # Replace with the actual path to your dataset
df = pd.read_csv(file_path)

# Group data by product_id and sum the repair costs
grouped_data = df.groupby(['product_id', 'defect_type']).sum().reset_index()

# Sort the grouped data by defect_type to group colors together
grouped_data = grouped_data.sort_values(by='defect_type')

# Calculate the total cost for each defect_type
total_costs = df.groupby('defect_type')['repair_cost'].sum()

# Prepare the values for the treemap
sizes = grouped_data['repair_cost']
labels = [
    f"{row['product_id']}" if row['repair_cost'] > sizes.mean() else ""  # Show label only for larger rectangles
    for _, row in grouped_data.iterrows()
]
# Map colors and fill NaN values with a default color (e.g., 'gray')
color_map = {'Structural': 'red', 'Functional': 'lightcoral', 'Cosmetic': 'lightpink'}
colors = grouped_data['defect_type'].map(color_map).fillna('gray')

# Create the figure and adjust layout to fit the legend and title above
fig, ax = plt.subplots(figsize=(13, 8))
plt.subplots_adjust(top=0.75)  # Add space at the top for the legend

# Display the title
plt.suptitle("Treemap of Repair Costs by Product ID (Color by Defect Type)", fontsize=16, y=0.95)

# Create the legend above the main graph
legend_x = 0.1  # x-position for the legend box (centered)
legend_y = 1.2  # y-position for the legend box (above the main graph)
for index, (defect_type, total_cost) in enumerate(total_costs.items()):
    color = color_map.get(defect_type, 'gray')
    rect_x = legend_x
    rect_y = legend_y - index * 0.08
    plt.gcf().add_artist(plt.Rectangle((rect_x, rect_y), 0.02, 0.02, color=color, transform=ax.transAxes))
    ax.text(rect_x + 0.03, rect_y, f"Total costs of {defect_type.lower()} defects: ${total_cost:,.2f}", 
            fontsize=10, verticalalignment='center', transform=ax.transAxes)

# Create the treemap with smaller product_id labels
squarify.plot(sizes=sizes, label=labels, color=colors, alpha=0.7, ax=ax, text_kwargs={'fontsize': 8})
for rect in ax.patches:  # Iterate through all the patches (rectangles)
    rect.set_linewidth(0.5)  # Set a thinner border width
    rect.set_edgecolor("gray")  # Set a lighter border color

ax.axis('off')  # Turn off axis
plt.savefig("treemap_repair_costs.png", dpi=300, bbox_inches='tight')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Load your dataset
file_path = 'defects_data.csv'  # Replace with the actual path to your dataset
df = pd.read_csv(file_path)

# Define a custom order for severity
severity_order = ['Critical', 'Moderate', 'Minor']

# Group data by defect_type and severity and sum the repair costs
grouped_data = df.groupby(['defect_type', 'severity'])['repair_cost'].sum().reset_index()

# Sort the data first by defect_type, then by severity using the custom order
grouped_data['severity'] = pd.Categorical(grouped_data['severity'], categories=severity_order, ordered=True)
grouped_data = grouped_data.sort_values(by=['defect_type', 'severity'])

# Prepare the labels and values for the plot
labels = [
    f"{row['defect_type']} ({row['severity']})\n${row['repair_cost']:.2f}"
    for _, row in grouped_data.iterrows()
]
values = grouped_data['repair_cost'].values

# Define the colors
colors = [
    '#324079', '#556AC1', '#9FB2FE',  # Functional group: light to dark blue
    '#795C32', '#C1A655', '#FEDE9F',  # Structural group: light to dark yellow
    '#327943', '#55C161', '#9FFEAF'  # Cosmetic group: light to dark green
]

# Number of bars
num_bars = len(values)

# Create a figure and polar axis
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw={'projection': 'polar'})

# Compute the width and theta for each bar
theta = np.linspace(0, 2 * np.pi, num_bars, endpoint=False)
width = 2 * np.pi / num_bars

# Plot each bar
for i in range(num_bars):
    ax.bar(
        theta[i],  # Angle
        values[i],  # Length of the bar
        width=width,  # Width of the bar
        color=colors[i % len(colors)],  # Color of the bar
        edgecolor='white',  # Border color
        linewidth=2  # Border width
    )

# Set the labels with the repair cost included
ax.set_xticks(theta)
ax.set_xticklabels(labels, fontsize=10, ha='center', va='center')

# Remove the outer border
ax.spines['polar'].set_visible(False)

# Hide the grid and y-ticks
ax.grid(False)
ax.set_yticklabels([])
ax.set_yticks([])

# Add a title to the graph
ax.set_title("Repair Costs by Defect Type and Severity", va='bottom', fontsize=14)
plt.savefig("repair_costs_by_defect_type_and_severity.png", dpi=300, bbox_inches='tight')
# Display the plot
plt.show()
```

```python
import pandas as pd
import networkx as nx
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches

# Load your actual dataset
file_path = 'defects_data.csv'  # Replace with the actual path to your dataset
df = pd.read_csv(file_path)

# Create the graph
G = nx.Graph()

# Add edges between product_id and defect_type
for _, row in df.iterrows():
    G.add_edge(f"{row['product_id']}", row['defect_type'])  # Only display product ID

# Generate the initial layout
pos = nx.spring_layout(G, seed=42, k=0.3)

# Manually adjust the positions of the main defect type nodes
pos['Functional'] = [0, 1]  # Move Functional up
pos['Structural'] = [-1, -0.5]  # Move Structural to the left
pos['Cosmetic'] = [1, -0.5]  # Move Cosmetic to the right

# Function to determine the color of the product nodes based on connections
def get_product_node_color(node):
    connections = set(G.neighbors(node))
    has_structural = 'Structural' in connections
    has_functional = 'Functional' in connections
    has_cosmetic = 'Cosmetic' in connections

    if has_structural and has_functional and has_cosmetic:
        return '#FA9794'  # Connected to all three
    elif has_structural and has_functional:
        return '#D4AAF0'  # Connected to Structural and Functional
    elif has_cosmetic and has_functional:
        return '#FEDE9F'  # Connected to Cosmetic and Functional
    elif has_cosmetic and has_structural:
        return '#B18B3F'  # Connected to Cosmetic and Structural
    else:
        return 'lightblue'  # Default color for product nodes

# Prepare node sizes and colors
node_sizes = [300 if node.isdigit() else 1500 for node in G.nodes()]  # Smaller size for product nodes
node_colors = [
    get_product_node_color(node) if node.isdigit() else
    'lightcoral' if node == 'Structural' else
    'lightgreen' if node == 'Cosmetic' else
    'lightsteelblue'
    for node in G.nodes()
]

# Draw the network graph
plt.figure(figsize=(10, 7))
nx.draw(
    G, pos, with_labels=True,
    node_color=node_colors,
    edge_color='gray',
    node_size=node_sizes,  # Use the prepared node sizes
    font_size=8,  # Set a fixed font size
    font_color='black'
)
plt.title("Network Graph: Product IDs and Defect Types")

# Create a legend
legend_handles = [
    mpatches.Patch(color='#FA9794', label='Connected to Structural, Functional, and Cosmetic'),
    mpatches.Patch(color='#D4AAF0', label='Connected to Structural and Functional'),
    mpatches.Patch(color='#FEDE9F', label='Connected to Cosmetic and Functional'),
    mpatches.Patch(color='#B18B3F', label='Connected to Cosmetic and Structural'),
    mpatches.Patch(color='lightblue', label='Connected to only one defect type')
]
plt.legend(handles=legend_handles, loc='upper right', fontsize=8)
plt.savefig("network_graph.png", dpi=300, bbox_inches='tight')
plt.show()
```
